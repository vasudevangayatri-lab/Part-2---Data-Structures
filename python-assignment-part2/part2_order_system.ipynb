### Task 1 — Explore the Menu

```python
# Grouping orders per ategories 
categories = []
for details in menu.values():
    cat = details["category"]
    if cat not in categories:          
        categories.append(cat)

# Printing orders per category
for category in categories:
    print(f"\n{'=' * 5} {category} {'=' * 5}")
    for item_name, details in menu.items():
        if details["category"] == category:
            status = "[Available]" if details["available"] else "[Unavailable]"
            print(f"  {item_name:<18} ₹{details['price']:>6.2f}   {status}")
```

```python
# Part B-Menu statistics

total_items = len(menu)
available_items = sum(1 for d in menu.values() if d["available"])
most_expensive = max(menu, key=lambda item: menu[item]["price"])
budget_items = {name: d for name, d in menu.items() if d["price"] < 150}
print(f"Total menu items    : {total_items}")
print(f"Available items     : {available_items}")
print(f"Most expensive item : {most_expensive}  (₹{menu[most_expensive]['price']:.2f})")
print("\nItems priced under ₹150:")
for name, d in budget_items.items():
    print(f"  {name:<18} ₹{d['price']:.2f}")
```

### Task 2 — Cart Operations

```python
# TASK 2  Order taking functions for cart operations

cart = []   

def add_to_cart(item_name, quantity):
    """Add an item to cart or increase quantity if already present."""
    
    if item_name not in menu:
        print(f"  Sorry'{item_name}' does not exist in the menu.")
        return
    if not menu[item_name]["available"]:
        print(f"  Sorry'{item_name}' is currently unavailable.")
        return
  
    for entry in cart:
        if entry["item"] == item_name:
            entry["quantity"] += quantity   
            print(f"  Updated '{item_name}' quantity to {entry['quantity']}.")
            return
   
    cart.append({"item": item_name, "quantity": quantity, "price": menu[item_name]["price"]})
    print(f"  Added '{item_name}' x{quantity} to cart.")


def remove_from_cart(item_name):
    """Remove an item from the cart by name."""
    for entry in cart:
        if entry["item"] == item_name:
            cart.remove(entry)
            print(f"  Removed '{item_name}' from cart.")
            return
            print(f"  sorry '{item_name}' is not in the cart.")


def update_quantity(item_name, new_quantity):
    """Update the quantity of an item already in the cart."""
    for entry in cart:
        if entry["item"] == item_name:
            entry["quantity"] = new_quantity
            print(f"  Right'{item_name}' quantity updated to {new_quantity}.")
            return
    print(f"  Sorry '{item_name}' is not in the cart.")


def print_cart(label="Current Cart"):
    """Display cart contents in a readable format."""
    print(f"\n  [{label}]")
    if not cart:
        print("  (empty)")
    else:
        for entry in cart:
            line_total = entry["quantity"] * entry["price"]
            print(f"  • {entry['item']:<18} x{entry['quantity']}  ₹{line_total:.2f}")
    print()


print(" << Cart functions defined. >>")
```

```python
# Simulating order sequence


print("═" * 42)
print("         CART SIMULATION")
print("═" * 42)

# Adding Paneer Tikka × 2
print("\n Add 'Paneer Tikka' x2")
add_to_cart("Paneer Tikka", 2)
print_cart()

# Adding Gulab Jamun × 1
print("Add 'Gulab Jamun' x1")
add_to_cart("Gulab Jamun", 1)
print_cart()

# Adding Paneer Tikka × 1  Should update quantity to existing + 1
print("Add 'Paneer Tikka' x1  (expect quantity existing +1)")
add_to_cart("Paneer Tikka", 1)
print_cart()

# Trying to add 'Mystery Burger' - not in menu
print("Add 'Mystery Burger' x1  (should fail — not in menu)")
add_to_cart("Mystery Burger", 1)
print_cart()

# Trying to add 'Chicken Wings'  - there in menu but unavailable
print("Add 'Chicken Wings' x1  (should fail — unavailable)")
add_to_cart("Chicken Wings", 1)
print_cart()

# Removing 'Gulab Jamun' from cart
print("Remove 'Gulab Jamun'")
remove_from_cart("Gulab Jamun")
print_cart()
```

```python
# Final Order Summary with GST @ 5%

GST_RATE = 0.05 

subtotal = sum(entry["quantity"] * entry["price"] for entry in cart)
gst      = subtotal * GST_RATE
total    = subtotal + gst

print("\n" + "=" * 38)
print("         Order Summary")
print("=" * 38)
for entry in cart:
    line_total = entry["quantity"] * entry["price"]
    print(f"{entry['item']:<20} x{entry['quantity']}  ₹{line_total:>7.2f}")
print("-" * 38)
print(f"{'Subtotal:':<28} ₹{subtotal:>7.2f}")
print(f"{'GST (5%):':<28} ₹{gst:>7.2f}")
print(f"{'Total Payable:':<28} ₹{total:>7.2f}")
print("=" * 38)
```

### Task 3 — Inventory Tracker with Deep Copy

```python
import copy

inventory_backup = copy.deepcopy(inventory)

print("━" * 48)
print(" Deep Copy Demonstration")
print("━" * 48)
print("\n- Changing inventory['Paneer Tikka']['stock'] to 999 ...")
inventory["Paneer Tikka"]["stock"] = 999

print(f"  inventory['Paneer Tikka']['stock']        = {inventory['Paneer Tikka']['stock']}")
print(f"  inventory_backup['Paneer Tikka']['stock'] = {inventory_backup['Paneer Tikka']['stock']}")
print("   Backup is UNAFFECTED — deep copy confirmed.")

inventory["Paneer Tikka"]["stock"] = 10
print("\n- Restored inventory['Paneer Tikka']['stock'] back to 10.")
```

```python
# Order fulfilment — deduct cart

print("\n━" * 14)
print(" Order Fulfilment — Stock Deduction")
print("━" * 14)
print(f"\n  Cart to fulfil: {[e['item'] + ' x' + str(e['quantity']) for e in cart]}\n")

for entry in cart:
    item     = entry["item"]
    qty_needed = entry["quantity"]
    available_stock = inventory[item]["stock"]

    if qty_needed > available_stock:
       
        print(f"  Warning: Insufficient stock for '{item}'. "
              f"Needed {qty_needed}, available {available_stock}. Deducting {available_stock}.")
        inventory[item]["stock"] = 3
    else:
        inventory[item]["stock"] -= qty_needed
        print(f"  Deducted {qty_needed} of '{item}'. Remaining stock: {inventory[item]['stock']}")

print("\n── Alert to ReOrder ──")
alert_found = False
for item, data in inventory.items():
    if data["stock"] <= data["reorder_level"]:
        print(f"  Reorder Alert: {item} — Only {data['stock']} unit(s) left "
              f"(reorder level: {data['reorder_level']})")
        alert_found = True
if not alert_found:
    print("  All items are adequately stocked.")
```

```python
# Compare current inventory vs backup

print("\n━" * 14)
print(" Inventory vs Backup Comparison")
print("━" * 14)
print(f"{'Item':<18} {'Available Stock':>12} {'Backup Stock':>14}")
print("-" * 46)
for item in inventory:
    live   = inventory[item]["stock"]
    backup = inventory_backup[item]["stock"]
    diff   = "  - changed" if live != backup else ""
    print(f"  {item:<18} {live:>8}      {backup:>8}  {diff}")

print("\n Deep copy protected the original inventory values.")
```

### Task 4 — Daily Sales Log Analysis

```python
# To show Revenue per day + best-selling of the day

def print_revenue_table(log, label="Revenue per Day"):
    """Print daily revenue and return the best-selling day."""
    print(f"\n{'━' * 38}")
    print(f"  {label}")
    print(f"{'━' * 38}")
    print(f"  {'Date':<14} {'Orders':>7} {'Revenue':>10}")
    print("-" * 38)

    daily_totals = {}  
    for date, orders in log.items():
        day_total = sum(order["total"] for order in orders)
        daily_totals[date] = day_total
        print(f"  {date:<14} {len(orders):>7}  ₹{day_total:>8.2f}")

    best_day = max(daily_totals, key=lambda d: daily_totals[d])
    print("-" * 38)
    print(f"  Best-selling day: {best_day}  (₹{daily_totals[best_day]:.2f})")
    print(f"{'━' * 38}\n")
    return best_day

best_day_initial = print_revenue_table(sales_log, "Revenue per Day (Initial)")
```

```python
# Most ordered item

item_order_count = {}   

for date, orders in sales_log.items():
    for order in orders:
        for item in order["items"]:
            item_order_count[item] = item_order_count.get(item, 0) + 1

most_ordered = max(item_order_count, key=lambda i: item_order_count[i])

print("Item Order Frequency:")
for item, count in sorted(item_order_count.items(), key=lambda x: -x[1]):
    print(f"  {item:<18} {count:>3} order(s)")

print(f"\n Most ordered item: '{most_ordered}' — appeared in {item_order_count[most_ordered]} order(s).")
```
